```python
import pandas as pd
from pathlib import Path

BASE = Path.cwd().parents[0]
PLAYERS_DIR = BASE / "data" / "processed" / "players"

files = sorted(PLAYERS_DIR.glob("*_players_all.csv"))
print("üì¶ –ù–∞–π–¥–µ–Ω—ã —Ñ–∞–π–ª—ã:")
for f in files:
    print(" -", f.name)

dfs = []
for f in files:
    df = pd.read_csv(f)
    df["league"] = f.stem.split("_")[0]
    dfs.append(df)

players_all = pd.concat(dfs, ignore_index=True)
print(f"\n‚úÖ players_all shape: {players_all.shape}")

dups = players_all.duplicated(subset=["season","player","squad","league"]).sum()
print(f"üîé –î—É–±–ª–∏–∫–∞—Ç—ã: {dups}")

players_all.to_csv(PLAYERS_DIR / "players_all.csv", index=False)
print(f"üìÅ –°–æ—Ö—Ä–∞–Ω–µ–Ω–æ: {PLAYERS_DIR / 'players_all.csv'}")
```
