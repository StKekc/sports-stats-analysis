```python
from pathlib import Path
import pandas as pd

BASE = Path("..").resolve()
ROOT = BASE / "data" / "raw" / "fbref"

seasons = ["epl_2019-2020","epl_2020-2021","epl_2021-2022","epl_2022-2023","epl_2023-2024","epl_2024-2025"]

rows = []
for s in seasons:
    d = ROOT / s
    st = pd.read_csv(d / "standings.csv")
    ts = pd.read_csv(d / "team_standard_stats.csv")
    fx = pd.read_csv(d / "schedule_results.csv")

    rows.append({
        "season": s.replace("epl_",""),
        "standings_rows": len(st),
        "team_stats_rows": len(ts),
        "fixtures_rows": len(fx),
        "standings_cols": len(st.columns),
        "team_stats_cols": len(ts.columns),
        "fixtures_cols": len(fx.columns),
    })

check = pd.DataFrame(rows)
check
```
